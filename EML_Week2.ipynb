```python
df = pd.read_csv('train.csv', usecols=['Age','Fare','Survived'])
df.head()
```

```python
df.isnull().mean()
```

```python
def input_nan(df,variable,median):
  df[variable+'median']= df[variable].fillna(median)

input_nan(df,'Age',df.Age.median())
df['age_mean']= df['Age'].fillna(df.Age.mean())
df.head()
```

```python
df['Age'].std()
```

```python
df.Age.std(),df.Agemedian.std()
```

```python
df.Age.std(), df.age_mean.std()
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')

matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (12,6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'

df.Age.value_counts()
```

```python
fig = plt.figure()
axes = fig.add_subplot(111)

df.Age.plot(kind='kde') #kernel density estimate
df.Agemedian.plot(kind='kde',color='red',ax=axes)
df.age_mean.plot(kind='kde',color='green',ax=axes)

lines,label = axes.get_legend_handles_labels()
axes.legend(lines,label,loc='best')
```
