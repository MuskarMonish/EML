# Implement binary Classification using ID3 Decision Tree Algorithm- Medical ApplicationDomain

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from matplotlib import pyplot as plt
col_names = ['Pregnancies', 'Glucose', 'BP', 'SkinThickness', 'Insulin', 'BMI', 'Pedigree', 'Age', 'Outcome']
pima = pd.read_csv("pima-indians-diabetes.csv", names = col_names)
pima.head()
```

```python
import seaborn as sns
corr = pima.corr()
ax= sns.heatmap(corr, vmin= -1, vmax=1, center= 0, cmap= sns.diverging_palette(20,220,n=200), square= True)
ax.set_xticklabels(ax.get_xticklabels(), rotation= 45, horizontalalignment= 'right');
```

```python
#feature selection
feature_cols=['Pregnancies', 'Insulin', 'BMI', 'Age','Glucose', 'BP', 'Pedigree']
X = pima[feature_cols]
y = pima.Outcome

#split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=1)
#build model
classifier = DecisionTreeClassifier()
classifier = classifier.fit(X_train, y_train)
#predict
y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
print(confusion_matrix(y_test, y_pred))
```

```python
#accuracy
print('Accuracy:', metrics.accuracy_score(y_test, y_pred))
```

```python
#import tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure()
plot_tree(classifier, filled = True)
plt.show()
```
